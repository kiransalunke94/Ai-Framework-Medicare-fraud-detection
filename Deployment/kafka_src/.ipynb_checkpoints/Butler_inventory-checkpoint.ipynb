```python
from scipy.stats import uniform,norm,rv_discrete
import pandas as pd
```

```python
trails = 300
```

```python
profit_unit = 50
holding_unit = 15
shortage_unit = 30
```

```python
#Replenishment level
stock = 140
```

```python
#Demand
mean = 100
std = 20
```

```python
#Creating dataframe with demand

df = pd.DataFrame({"Demand": norm.rvs(loc=mean,scale=std,size=trails)})
df["Demand"] = df["Demand"].round(2)
df.head(5)
```

```python
#Creating sales column with min value of stock and demand

df["Sales"] = [min(stock,i) for i in df["Demand"]]
df.head(5)
```

```python
#Adding gross profit as multi of profit per unit * sales

df["Gross_profit"] = [profit_unit*i for i in df["Sales"]]
df.head(5)
```

```python
def Holding_cost_fun(data):
    if data < stock:
        cost = holding_unit*(stock-data)
    else:
        cost = 0
    return cost
```

```python
#Holding cost : Inventory not sold, cost of storing the fans
#(inventory-demand)*holding_unit

df["Holding_cost"] = [Holding_cost_fun(i) for i in df["Demand"]]
df.head(5)
```

```python
def shortage_cost_fun(data):
    if data > stock:
        cost  = (data-stock)*shortage_unit
    else:
        cost = 0
    return cost
```

```python
#Shortage cost : incase demand is more than inventory
#(demand-inventory) * shortage cost
df["Shortage_cost"] = [shortage_cost_fun(i) for i in df["Demand"]]
df.head(5)
```

```python
#Net profit
#Profit - Holding cost - shortage cost

df["Net_profit"] = [df.loc[i,"Gross_profit"] - df.loc[i,"Holding_cost"] - df.loc[i,"Shortage_cost"] for i in range(0,len(df))]
df.head()
```

```python
#For Replenishment level 120

print("mean profit : ",df["Net_profit"].mean().round(2))
print("Std : ",df["Net_profit"].std().round(2))
print("Min profit : ", min(df["Net_profit"]))
print("Max profit : ", max(df["Net_profit"]))
print("Service level : ",(df["Sales"].sum()/df["Demand"].sum()*100).round(2))
```
