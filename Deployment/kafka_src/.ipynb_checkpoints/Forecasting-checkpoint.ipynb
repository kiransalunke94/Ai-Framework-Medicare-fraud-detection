```python
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_excel("Closing Stock Prices.xlsx",skiprows=2)
```

```python
df.head()
```

```python
def roll_mean(df,name,out_name,window_size,shift_unit):
    df[out_name] = df[name].rolling(window=window_size).mean().shift(shift_unit).round(2)
    return(df)
```

```python
roll_mean(df,"IBM","IBM_roll_2",2,1)
roll_mean(df,"IBM","IBM_roll_3",3,1)
roll_mean(df,"IBM","IBM_roll_4",4,1)
roll_mean(df,"INTC","INTC_roll_2",2,1)
roll_mean(df,"INTC","INTC_roll_3",3,1)
roll_mean(df,"INTC","INTC_roll_4",4,1)
roll_mean(df,"CSCO","CSCO_roll_2",2,1)
roll_mean(df,"CSCO","CSCO_roll_3",3,1)
roll_mean(df,"CSCO","CSCO_roll_4",4,1)
roll_mean(df,"GE","GE_roll_2",2,1)
roll_mean(df,"GE","GE_roll_3",3,1)
roll_mean(df,"GE","GE_roll_4",4,1)
roll_mean(df,"DJ Industrials \nIndex","DJ_roll_2",2,1)
roll_mean(df,"DJ Industrials \nIndex","DJ_roll_3",3,1)
roll_mean(df,"DJ Industrials \nIndex","DJ_roll_4",4,1)
```

```python
df.dropna(how="any",inplace=True)
```

```python
def metrics()    
    print(mean_absolute_percentage_error(df["Units Sold"], df["ma_4"])) #MAPE
    print(mean_squared_error(df["Units Sold"], df["ma_4"], squared=False)) #RMSE
    print(mean_squared_error(df["Units Sold"], df["ma_4"], squared=True)) #MSE
    print(mean_absolute_error(df["Units Sold"], df["ma_4"])) #MAD
```
